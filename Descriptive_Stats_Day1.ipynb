```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('titanic')
df.head()
```

```python
# mean/ median/mode
print(df['age'].mean())# Mean ≈ Median → data may be approximately symmetric Mean > Median → usually right-skewed (positive skew) Mean < Median → usually left-skewed (negative skew)
print(df['age'].median())
print(df['embark_town'].mode())
```

```python
#range, variance, standard deviation,coefficient of variance
print(df['age'].max()-df['age'].min())
print(df['age'].var())
print(df['age'].std())
print(df['age'].std()/df['age'].mean()*100)# from age and fare coefficient of variance we can say age feature values are more close to mean where as fare feature values are more far from mean(more spread out)
print(df['fare'].std()/df['fare'].mean()*100)
```

```python
count=df['sex'].value_counts()
plt.bar(count.index,count.values)
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()
```

```python
plt.pie(count.values,labels=count.index,autopct='%1.1f%%')
plt.show()
```

```python
df['age'].cov(df['fare'])# positive means if age increases fare also increase
# negative means if age increases fare decrease
#0 means independent

#In machine learning  the feature with highest covariance with target is the best feature to use in the model and the feature with low cow covariance between each other is the best feature to use in the model as if 2 feature is having high covariance means both tels almost same information about target variable and if 2 feature is having low covariance means both feature is having different information about target variable
```
